# Nested Conditional Statement

```python
#Login Funtion demo - riduan.com
"""
Actual Credemtials:

Email: riduan@gmail.com
Password: riduan46
"""
email = input("Enter your email: ")
password = input("Enter your password: ")

actua_email = "riduan@gmail.com"
actual_password = "riduan46"

if email == actua_email and password == actual_password:
    print("Login Successful")
elif email == actua_email and password != actual_password:
    print("Please Try again")
    password = input("Enter your passeword again: ")
    if password == actual_password:
        print("Success!")
    else:
        print("Wrong")
else:
    print("Imvalid Credentials!")
```
