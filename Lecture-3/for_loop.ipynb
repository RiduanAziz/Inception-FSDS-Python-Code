```python
for i in range(1,5):
    for j in range(1,5):
        print(i,j)
```

# Excercise: print triangle coordinates

```python
row = int(input("Enter the number of rows: "))

for i in range(1, row+1):
    for j in range(1, i+1):
        print("*", end = " ")
    print(" ")
```

# Loop Control Statement

- Break
- Continue
- Pass

```python
# break

for i in range(1,11):
    if (i == 5):
        break
    print(i)
```

```python
name = ["ALex", "Mamun", "Bappy", "Ridoy", "Debashish"]

for i in name:
    if i == "Bappy":
        print(i)
        break
    
```

```python
# Continue

for i in range(1,10):
    if (i == 5):
        continue
    print(i)
```

```python
# Pass

for i in range(1,10):
    pass
```
