## Creating Stings

```python
s = 'hello'
s = "hello"
# Multiline strings
s = '''hello'''
s = """hello"""
s = str('hello')
print(s)
```

```python
a = 34
str(a)
```

```python
text = "it's my bike"
print(text)
```

# Accessing Strings

```python
# Positive Indexing

s = "Welcome to python"
```

```python
print(s[0])
```

```python
print(s[11])
```

```python
print(s[50])
```

```python
len(s)
```

```python
# Negative Indexing

s = "Welcome to python"
```

```python
print(s[-1])
```

```python
print(s[-11])
```

```python
# string slicing

s = "Welcome to python"
```

```python
s[0:7]
```

```python
s[11:17]
```

```python
print(s[11:17])
```

```python
print(s[11:])
```

```python
print(s[0:7:1])
```

```python
print(s[0:7:2])
```

```python
print(s[6::-1])
```

```python
print(s[6::-2])
```
