# Making our own for loop

```python
def amar_for_loop(iterable):
    # fetch the iterator
    iter_obj = iter(iterable)

    while True:
        try:
            # fetch the next item
            print(next(iter_obj))
            
        except StopIteration:
            # if StopIteration is raised, break from loop
            break
```

```python
a= [1,2,3,4,5]
b = (1,2,4,5)
c = {1,2,3,4,5,7,9}
d = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5}
e = range(1, 6)
```

```python
amar_for_loop(a)
```

```python
amar_for_loop(b)
```

```python
amar_for_loop(c)
```

```python
amar_for_loop(d)
```

```python
amar_for_loop(e)
```

# Generator

```python
def gen():
    yield "First Statement"
    yield "Second Statement"
    yield "Third Statement"
```

```python
gen = gen()

for i in gen:
    print(i)
```
