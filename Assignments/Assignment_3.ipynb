# 1. Inventory Price Calculator

Real-world use: Used in e-commerce dashboards for price summaries.

Task:

Write a function calculate_inventory_summary (prices) that:
- receives a list of product prices
- returns total_cost and average_cost

Example:

Input: [120, 250, 399, 150]

Output: (919, 229.75)

```python
def calculate_inventory_summary(prices):
    if prices:
        total_cost = sum(prices)
        average_cost = total_cost / len(prices)
        return total_cost, average_cost
    else:
        return 0, 0

user_input = input("Enter the prices of items in inventory: ")
if user_input == "":
    prices = []
else:
    prices = list(map(float, user_input.split(",")))
result = calculate_inventory_summary(prices)
print(result)
```

# 2. Unique Visitor Tracker

Real-world use: Website analytics tools calculating daily active users.

Task:

Write a function count_unique_visitors(visitor_list) that:
- takes a list of visitor IDs
- returns the number of unique visitors

Example:

Input: [101, 205, 101, 310, 205, 550]

Output: 4 (The unique IDs are 101, 205, 310, 550)

```python
def count_unique_visitors(visitor_list):
    if visitor_list:
        unique_visitors = set(visitor_list)
        unique_visitors = sorted(unique_visitors, reverse=False)
        return  tuple(unique_visitors)
    else:
        return ()

user_input = input("Enter the Visitor IDs: ")
if user_input == "":
    visitor_list = []
else:
    visitor_list = list(map(int, user_input.split(",")))
result = count_unique_visitors(visitor_list)
if result == ():
    print("0, No visitors")
else:
    ids = ", ".join(map(str, result))
    print(f"{len(result)} (The unique IDs are {ids})")
```

# 3. Product Category Counter

Real-world use: Inventory/category-based analytics.

Task:

Write a function count_categories(products) that:
- takes a list of product names
- returns a dictionary counting each product type.

Example:

Input: ["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]

Output: {"Laptop": 3, "Smartphone": 2, "Monitor": 1, "Headphones": 1}

```python
def count_categories(products):
    if products:
        product_count = {}               
        for product in products:
            if product in product_count:
                product_count[product] += 1
            else:
                product_count[product] = 1
        return product_count
    else:
        return {}
    
user_input = input("Enter the Prodcut Names: ")

if user_input == "":
    products = []
else:
    products = user_input.split(", ")

result = count_categories(products)
if result:
    print(result)
else:
    print("No Product Entered")
```

# 4. Temperature Analytics Real-world use: 

Health tech temperature screening systems. 

Task: 
Write a function analyze_temperatures(temp_list) that:
- separates high (> 38°C) and normal temps
- returns two lists: (high_list, normal_list)

Example: 

Input: [36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]

Output:
High List (>38°C): [39.1, 40.5]

Normal List (<38°C): [36.5, 38.0, 37.2, 38.0, 37.9]

Full Output Tuple: ([39.1, 40.5], [36.5, 38.0, 37.2, 38.0, 37.9])

```python
def analyze_temperatures(temp_list):
    if temp_list:
        high_list = []
        normal_list = []
        for temp in temp_list:
            if temp > 38:
                high_list.append(temp)
            elif temp <= 38:
                normal_list.append(temp)
        return high_list, normal_list
    else:
        return {}
    
user_input = input("Enter the Temperatures: ")

if user_input == "":
    temp_list = []
else:
    temp_list = list(map(float, user_input.split(", ")))

high_temps, normal_temps = analyze_temperatures(temp_list)

if temp_list:
    print("High List (>38°C):", high_temps)
    print("Normal List (<38°C):", normal_temps)
    tuple_list = (high_temps, normal_temps)
    print("Full Output Tuple:", tuple_list)
else:
    print("No Temperatures Entered")
```

# 6. Student Grading System

Real-world use: School result automation.

Task:

Write a function calculate_grade(score) that:

- returns "A+", "A", "A-", "F" based on numeric marks

Write another function grade_students(student_dict) to:
- take a dictionary of student:score
- return student:grade dictionary

Example:

Input (Scores Dictionary): {"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69}

Output (Grades Dictionary): {"Alice": "A+", "Bob": "A", "Charlie": "F", "David": "A-"}

- Input : Alice, 85, Bob, 78, Charlie, 32, David, 69

```python
def calculate_grade(score):
    if 80 <= score <= 100:
        return 'A+'
    elif 70 <= score < 80:
        return 'A'
    elif 60 <= score < 70:
        return 'A-'
    elif 50 <= score < 60:
        return 'B'
    elif 40 <= score < 50:
        return 'C'
    elif 33 <= score < 40:
        return 'D'
    elif 0 <= score < 33:
        return 'F'
    else:
        return "Invalid score"
    

def grade_students(student_dict):
    if student_dict:
        grade_dict = {}
        for student, score in student_dict.items():
            grade = calculate_grade(score)
            grade_dict[student] = grade
        return grade_dict
    else:
        return {}
    
user_input = input("Enter student names and scores: ").split(", ")

student_dict = {}

for i in range(0, len(user_input), 2):
    key = user_input[i]
    value = int(user_input[i+1])
    student_dict[key] = value

result = grade_students(student_dict)
if result:
    print(result)
else:
    print("No Students Entered")
```

# 7. Discount Calculator

Real-world use: Online store discount engine.

Task:

Write a function apply_discounts(products) where:
- products is a list of tuples: (name, price, discount)
- return a list of updated prices after discount

Example:

Input (List of Tuples): [("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)]

Output (List of Prices): [64.00, 22.50, 285.00]

```python
def apply_discounts(products):
    if products:
        discounted_products = {}
        for product, price in products.items():
            if price > 80:
                discounted_price = price * 0.8 
            else:
                discounted_price = price * 0.95  # 5% discount
            discounted_products[product] = round(discounted_price, 2)
        return discounted_products
    else:
        return {}
```
