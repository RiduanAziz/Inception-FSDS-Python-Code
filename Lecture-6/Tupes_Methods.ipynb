# len/min/max/sorted/sum

```python
t = (1,2,3,40)

print(len(t))
print(min(t))
print(max(t))
print(sum(t))
print(sorted(t,reverse=True))
print(sorted(t,reverse=False))
```

### Difference between Lists and Tuples

- Syntax
- Mutability
- Speed
- Memory
- Built in functionality
- Error
- Usability

```python
import sys

L = list(range(1000))
T = tuple(range(1000))

print('List size',sys.getsizeof(L))
print('Tuple size',sys.getsizeof(T))
```

```python
a = [1,2,3]
b = a

a.append(4)
print(a)
print(b)
```

```python
print(id(a))
print(id(b))
```

```python
a = (1,2,3)
b = a

a = a + (4,)
print(a)
print(b)
```
