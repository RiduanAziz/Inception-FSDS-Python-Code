# Assignment 1: Temperature Check

Problem:

Write a program that takes the current temperature as input and prints:

- "It's too cold!" if below 10°C
- "It's cool outside" if between 10°C and 25°C
- "It's hot outside!" if above 25°C

```python
current_temp = int(input("Enter Current Temperature: "))

if current_temp < 10:
    print("It's too cold!")

elif current_temp >= 10 and current_temp <= 25:
    print("It's cool outside")

else:
    print("It's hot outside!")
```

# Assignment 2: ATM Withdrawal System

Problem:

Ask the user for their account balance and the amount they want to withdraw.
- If the withdrawal amount is greater than the balance → print "Insufficient Balance"
- If the withdrawal amount is less than or equal to balance → print "Transaction Successful" and show the remaining balance.

```python
balance = int(input("Enter Account Balance: "))
withd_amt = int(input("Enter Withdrawal Amount: "))

if withd_amt > balance:
    print("Insufficient Balance")

elif withd_amt <= balance:
    print("Transaction Successful")
    rem_amt = balance - withd_amt
    print("Remaining Balance is: ", rem_amt)
```

# Assignment 5: Restaurant Bill with Discount

Problem:

Take total bill amount from the user:
- If the bill is above 1000 → apply 10% discount
- Otherwise → no discount

Then print the final bill amount.

```python
bill = int(input("Input Bill Amount of the User : "))

if bill > 1000:
    final_bill = bill-(bill*0.10)
    print("Final Bill Amount with 10% Discount: ", final_bill)

else:
    print("Final Bill Amount: ", bill)
```

# Assignment 8: Leap Year Checker

Problem:

Take a year as input and check:

- If it is divisible by 400 → Leap Year
- Else if divisible by 4 but not by 100 → Leap Year
- Otherwise → Not a Leap Year

```python
year = int(input("Enter a year: "))

if year%400 == 0:
    print("Leap Year")

elif year%4 == 0 and year%100 != 0:
    print("Leap Year")

else:
    print("Not a Leap Year")
```

# Assignment 10: Electricity Bill Calculator

Problem:

Input number of units consumed and calculate bill:

- Up to 100 units → 5 BDT per unit
- 101–200 units → 7 BDT per unit
- Above 200 units → 10 BDT per unit

Print total payable amount.

```python
units = int(input("Enter the number of unit consumed: "))
bill1 = int(input("Enter Calculate Bill-1 (for Up to 100 units): "))
bill2 = int(input("Enter Calculate Bill-2 (for 101–200 units): "))
bill3 = int(input("Enter Calculate Bill-3 (for Above 200 units): "))


if units >= 0 and units <= 100:
    pay_amt = units * bill1
    print("Total Payable amount: ", pay_amt)

elif units >= 101 and units <= 200:
    amt1 = 100 * bill1
    amt2 = (units-100) * bill2
    pay_amt = amt1 + amt2
    print("Total Payable amount: ", pay_amt)

else:
    amt1 = 100 * bill1
    amt2 = 100 * bill2
    amt3 = (units-200) * bill3
    pay_amt = amt1 + amt2 + amt3
    print("Total Payable amount: ", pay_amt)
    
```
