# File Handling in Python

##### Types of data used for I/O:
- Text - '12345' as a sequence of unicode chars
- Binary - 12345 as a sequence of bytes of its binary equivalent

##### Hence there are 2 file types to deal with
- Text files - All program files are text files
- Binary Files - Images,music,video,exe files



### How File I/O is done in most programming languages

- Open a file
- Read/Write data
- Close the file

# reading from files

```python
f = open('files/sample.txt','r')
s = f.read()
print(s)
f.close()
```

```python
# reading upto n chars
f = open('files/sample.txt','r')
s = f.read(15)
print(s)
f.close()
```

```python
# readline() -> to read line by line
f = open('files/sample.txt','r')
print(f.readline(),end='')
print(f.readline(),end='')
f.close()
```

```python
f = open('files/sample.txt','r')

while True:
    data = f.readline()

    if data == "":
        break
    else:
        print(data, end= "")


f.close()
```

# File Handling using Context Manager (With)

### Using Context Manager (With)

- It's a good idea to close a file after usage as it will free up the resources
- If we dont close it, garbage collector would close it
- with keyword closes the file as soon as the usage is over

```python
with open("files/sample.txt", "w") as f:
    f.write("Hello there!")
```

```python
f.write("Bappy")
```

```python
# read()
with open("files/sample.txt", "r") as f:
    s = f.read()
    print(s)
```

```python
# readline()
with open("files/sample.txt", "r") as f:
    s = f.readline()
    print(s)
```

```python
# append mode
with open("files/sample.txt", "a") as f:
    s = f.write("\nMy name is Bappy")
```

```python
# moving within a file
with open("files/sample.txt", "r") as f:
    print(f.read(10))
```

```python
with open("files/sample.txt", "r") as f:
    f.seek(5)
    print(f.read())
```

```python
with open("files/sample.txt", "r") as f:
    f.seek(5)
    print(f.tell())
```

# Problems with working in text mode

- can't work with binary files like images
- not good for other data types like int/float/list/tuples

```python
# working with binary file
with open("files/Riduan.png", 'r') as f:
    print(f.read())
```

```python
# working with binary file
with open("files/Riduan.png", 'rb') as f:
    print(f.read())
```

```python
# working with binary file
with open("files/Riduan.png", 'rb') as f:
    original_file = f.read()
    
    with open("files/Riduan_copy.png", "wb") as wf:
        wf.write(original_file)
```

### Serialization and Deserialization

- **Serialization** - process of converting python data types to JSON format
- **Deserialization** - process of converting JSON to python data types

#### What is JSON?
The full form of JSON is JavaScript Object Notation.

image.png

```python
import json 

L = [1,2,3,4,5,6]

with open("files/test.json", "w") as f:
    json.dump(L, f)
```

```python
import json 

with open("files/test.json", "r") as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

```python
# dict
d = {
    'name':'Bappy',
     'age':23,
     'gender':'male'
}

with open("files/test.json", "w") as f:
    json.dump(d, f, indent=4)
```

# Pickling in Python

### Pickling
`Pickling` is the process whereby a Python object hierarchy is converted into a byte stream, and `unpickling` is the inverse operation, whereby a byte stream (from a binary file or bytes-like object) is converted back into an object hierarchy.

```python
def display_info():
    return "Hi my name is Bappy and I am a Programmer"
```

```python
d = display_info()
```

```python
# pickle dump
import pickle
with open('files/function.pkl','wb') as f:
  pickle.dump(d,f)
```

```python
# pickle load
import pickle
with open('files/function.pkl','rb') as f:
  p = pickle.load(f)
  print(p)
```
