# Super Keyword

```python
class Phone:
    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.__price = price
    
    def get_price(self):
        return self.__price

    def buy(self):
        print(f"You have bought {self.brand} {self.model} for ${self.__price}.")


    
    def display_info(self):
        print(f"Brand: {self.brand}, Model: {self.model}, Price: {self.__price}")

class SmartPhone(Phone):
    def buy(self):
        print(f"You have bought smartphone {self.brand} {self.model} for ${self.get_price()}.")
        super().buy()


```

```python
s = SmartPhone("Apple", "iPhone 13", "50000")
```

```python
s.buy()
```

# Types of Inheritance in OOP

- Single Inheritance
- Multilevel Inheritance
- Hierarchical Inheritance
- Multiple Inheritance

```python
# Multilevel Inheritance

#GrandFather
class Product:
    def review(self):
        print("This is a great product!")
    
#Father
class Phone(Product):
    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.__price = price

    def buy(self):
        print("Buying the phone...")

#Child
class SmartPhone(Phone):
    pass

s = SmartPhone("Apple", "iPhone 13", "50000")
s.review()
s.buy()
```

```python
# Hierarchical Inheritance
    
#Father
class Phone(Product):
    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.__price = price

    def buy(self):
        print("Buying the phone...")

#Child
class SmartPhone(Phone):
    pass

#Child
class FeaturePhone(Phone):
    pass

s = SmartPhone("Apple", "iPhone 13", "50000")
s.buy()

F= FeaturePhone("Nokia", "3310", "2000")
F.buy()
```

```python
# Multiple Inheritance

#Parent 1
class Camera:
    def click(self):
        print("Clicking a photo...")

#Parent 2
class MusicPlayer:
    def play_music(self):
        print("Playing music...")

#Child
class SmartPhone(Camera, MusicPlayer):
    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.__price = price

s = SmartPhone("Apple", "iPhone 13", "50000")
s.click()
s.play_music()
```

Polymorphism in OOP

```python
class Shape:
    def area(self, radius):
        print("Area of circle calculated.")
        return 3.14 * radius * radius
        
    
    def area(self, length, breadth):
        print("Area of rectangle calculated.")
        return length * breadth
        


s = Shape()
s.area(5, 10)
```

```python
class Shape:
    def area(self, radius, length, breadth):
        if breadth == None:
            print("Area of circle calculated.")
            return 3.14 * radius * radius
        else:
            print("Area of rectangle calculated.")
            return length * breadth
        

s = Shape()
s.area(5, 25, 20)
```

Abstraction in OOP

```python
# Abstract base class
from abc import ABC, abstractmethod

#Senior
class BankServr(ABC):
    def database(self):
        print("Connecting to database...")

    @abstractmethod
    def sequrity(self):
        print("Checking security...")

#Junior
class MobileApp(BankServr):
    def login(self):
        print("Logging in...")

    def sequrity(self):
        print("Mobile app security checked.")
```

```python
mob = MobileApp()
mob.database()
```
