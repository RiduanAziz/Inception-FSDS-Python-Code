Static Variables and Methods in OOP

```python
class AtmMachine:

    #static variable
    __customer_id = 0  # Class variable
    
    # Constructor (speecial function) --> super power of OOP
    def __init__(self):
        #print(id(self))
        self.pin = ""
        self.balance = 0
        #self.menu()
        AtmMachine.__customer_id += 1
        print(f"Your customer id is: {AtmMachine.__customer_id}")

    @staticmethod
    def get_customer_id():
        return AtmMachine.__customer_id
    

    def menu(self):
        user_input = input(
            """
            Hi, Welcome to ATM Machine
            How can I help you?
            1. Create Pin
            2. Change Pin
            3. Check Balance
            4. Withdraw Money
            5. Exit
            """
        )
        if user_input == "1":
            self.create_pin()
        elif user_input == "2":
            self.change_pin()
        elif user_input == "3":
            self.check_balance()
        elif user_input == "4":
            self.withdraw_money()
        elif user_input == "5":
            print("Thank you for using ATM Machine")
        else:
            print("Invalid Input")
    
    def create_pin(self):
        user_pin = input("Please enter your pin: ")
        self.pin = user_pin
        print("Pin created successfully")
        user_balance = float(input("Please enter your balance: "))
        self.balance = user_balance
        print("Balance added successfully")
        #self.menu()
    
    def change_pin(self):
        user_pin = input("Please enter your current pin: ")
        if user_pin == self.pin:
            new_pin = input("Please enter your new pin: ")
            self.pin = new_pin
            print("Pin changed successfully")
        else:
            print("Invalid Pin")
        #self.menu()

    def check_balance(self):
        user_pin = input("Please enter your pin: ")
        if user_pin == self.pin:
            print(f"Your balance is: {self.balance}")
        else:
            print("Invalid Pin")
        #self.menu()

    def withdraw_money(self):
        user_pin = input("Please enter your pin: ")
        if user_pin == self.pin:
            amount = float(input("Please enter the amount to withdraw: "))
            if amount <= self.balance:
                self.balance -= amount
                print(f"Please take your cash: {amount}")
                print(f"Your remaining balance is: {self.balance}")
            else:
                print("Insufficient Balance")
        else:
            print("Invalid Pin")
        #self.menu()
```

```python
class Customer:
    def __init__(self, name, gender, address):
        self.name = name
        self.gender = gender
        self.address = address
    
    def display_info(self):
        print(f"Name: {self.name}, Gender: {self.gender}, Address: {self.address._Address__city}, {self.address._Address__state}, {self.address._Address__country}.")

class Address:
    def __init__(self, city, state, country):
        self.__city = city
        self.__state = state
        self.__country = country

    @property
    def city(self):
        return self.__city

    @property
    def state(self):
        return self.__state

    @property
    def country(self):
        return self.__country
```

# Inheritance in OOP

Inheritance allows us to define a class that inehrits all the methods and properties from another class. Parent clas is the class being inherited from, also called base class. Child ckass is the class that inherits from another class, also called derived class.

- non private attribute
- non private methods

```python
class User: #Parent class
    def __init__(self):
        self.name = "Riduan"
        self.email = "riduan@example.com"

    def login(self):
        print(f"{self.name} logged in with email: {self.email}")

    def register_user(self):
        print(f"{self.name} registered with email: {self.email}")


class Student(User): #Child class
    def __init__(self):
        self.student_id = "12345"

    def enroll_course(self, course_name):
        print(f"{self.name} enrolled in course: {course_name}")

    def review_course(self, course_name):
        print(f"{self.name} reviewed course: {course_name}")

    def display_student_info(self):
        print(f"Student ID: {self.student_id}, Name: {self.name}, Email: {self.email}")


class Instructor(User): #Child class
    def __init__(self):
        self.instructor_id = "54321"

    def create_course(self, course_name):
        print(f"{self.name} created course: {course_name}")

    def reply_to_review(self, course_name):
        print(f"{self.name} replied to review for course: {course_name}")

    def display_instructor_info(self):
        print(f"Instructor ID: {self.instructor_id}, Name: {self.name}, Email: {self.email}")
```

```python
s = Student()
s.login()
s.name
```
