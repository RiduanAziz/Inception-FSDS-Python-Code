# OOP

```python
# using OOP

class Robot:

    def __init__(self, robot_number):
        self.robot_number = robot_number

    def movement(self):
        print(f"Its Robot {self.robot_number} is moving forward.")

    def fire_detection(self):
        print(f"Robot {self.robot_number} is detecting fire.")

    def wheel_control(self):
        print(f"Robot {self.robot_number} is controlling its wheels.")

    def ai_intelligence(self):
        print(f"Robot {self.robot_number} is using AI intelligence.")
```

```python
robot1 = Robot(robot_number=1)
```

```python
robot1.movement()
```

```python
robot2 = Robot(robot_number=2)
robot2.movement()
```

```python
robot1.ai_intelligence()
```
