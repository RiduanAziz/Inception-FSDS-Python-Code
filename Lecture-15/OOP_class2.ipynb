ATM machine - AtmMachine

```python
class AtmMachine:
    # Constructor (speecial function) --> super power of OOP
    def __init__(self):
        self.pin = ""
        self.balance = 0
        self.menu()

    def menu(self):
        user_input = input(
            """
            Hi, Welcome to ATM Machine
            How can I help you?
            1. Create Pin
            2. Change Pin
            3. Check Balance
            4. Withdraw Money
            5. Exit
            """
        )
        if user_input == "1":
            self.create_pin()
        elif user_input == "2":
            self.change_pin()
        elif user_input == "3":
            self.check_balance()
        elif user_input == "4":
            self.withdraw_money()
        elif user_input == "5":
            print("Thank you for using ATM Machine")
        else:
            print("Invalid Input")
    
    def create_pin(self):
        user_pin = input("Please enter your pin: ")
        self.pin = user_pin
        print("Pin created successfully")
        user_balance = float(input("Please enter your balance: "))
        self.balance = user_balance
        print("Balance added successfully")
        self.menu()
    
    def change_pin(self):
        user_pin = input("Please enter your current pin: ")
        if user_pin == self.pin:
            new_pin = input("Please enter your new pin: ")
            self.pin = new_pin
            print("Pin changed successfully")
        else:
            print("Invalid Pin")
        self.menu()

    def check_balance(self):
        user_pin = input("Please enter your pin: ")
        if user_pin == self.pin:
            print(f"Your balance is: {self.balance}")
        else:
            print("Invalid Pin")
        self.menu()

    def withdraw_money(self):
        user_pin = input("Please enter your pin: ")
        if user_pin == self.pin:
            amount = float(input("Please enter the amount to withdraw: "))
            if amount <= self.balance:
                self.balance -= amount
                print(f"Please take your cash: {amount}")
                print(f"Your remaining balance is: {self.balance}")
            else:
                print("Insufficient Balance")
        else:
            print("Invalid Pin")
        self.menu()



    
```
