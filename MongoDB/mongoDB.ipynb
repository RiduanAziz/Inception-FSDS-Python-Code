```python
DB_NAME = "covid19_db"
COLLECTION_NAME = "covid19_collection"
COLLECTION_URL = "<mongoDB connection_string goes here>"
```

```python
import pymongo

client = pymongo.MongoClient(COLLECTION_URL)
database = client[DB_NAME]
collection = database[COLLECTION_NAME]
```

## Storing Data in MongoDB

```python
records = collection.insert_many(data_dict)
```

## Fetching Data from MongoDB

```python
records = collection.find()
```
